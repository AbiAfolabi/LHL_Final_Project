```python
api_key = os.environ["YELP API KEY"]
```

```python
url = "https://api.yelp.com/v3/businesses/_zkBt4pyAluwbShtQVNS4A/reviews?limit=20&sort_by=yelp_sort"
```

```python
# Create dictionary for headers
headers = {"Accept": "application/json"}

# Add key with our API KEY
headers['Authorization'] = api_key
```

```python
# Send the request to Yelp:

result = requests.get(url, headers=headers)

result
```

```python
df_result = pd.json_normalize(result.json())

df_result
```
