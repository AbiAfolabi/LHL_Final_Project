## 2. Querying Yelp for African stores in Calgary using the term "african store"

```python
## specifying location of Yelp API key stored in computer's environmental variables. 

api_key = os.environ["YELP API KEY"]
```

```python
## specifying url to retrieve data from.

url = "https://api.yelp.com/v3/businesses/search?location=calgary&term=african%20store&radius=40000&sort_by=best_match&limit=20"
```

```python
# Create dictionary for headers
headers = {"Accept": "application/json"}

# Add key with our API KEY
headers['Authorization'] = api_key
```

```python
# Send the request to Yelp:

result = requests.get(url, headers=headers)

result
```

```python
print(result.json())
```

```python
df_result = pd.json_normalize(result.json())

df_result
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_businesses =pd.json_normalize(df_result['businesses'][0])

df_businesses
```
