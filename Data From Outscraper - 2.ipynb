## 1. Importing Required Libraries.

```python
## import pandas to read csv file in a dataframe

import pandas as pd
```

```python
## set display option to show all rows and columns in the dataframe

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
## read csv file into a dataframe

df = pd.read_csv('outscraper_data2.csv', delimiter=',',encoding= 'unicode_escape')

df.head()
```

```python
## check the dimension (rows and columns) of the dataframe. 

df.shape
```

```python
## check for datatype in each column, rows with complete and incomplete data.

df.info()
```

## 2. Data Cleaning

#### i) Split query column to identify city where the reviews were collected.

```python
## split query column by delimeter ','

df_query = df['query'].str.split(',', expand=True)
df_query
```

```python
## split query[1] column by delimeter ' ' to have Calgary as a separate column.

df_query = df_query[1].str.split(' ', expand = True) 

df_query.head()
```

```python
## Merge the column for city df_query[1] with the original df

df_new = pd.concat([df_query[1], df], axis = 1)

df_new.head()
```

#### ii) After inspecting the dataframe, I will be removing columns not necessary for the NLP project at hand.

```python
df_resized = df_new[[1,'name', 'reviews', 'rating','review_id', 'review_text', 'review_rating', 'review_datetime_utc']]

df_resized.head()
```

#### iii) Renaming column '1' to city

```python
df_resized = df_resized.rename(columns = {1:'city'})

df_resized.head()
```

```python
## checking all city names retrieved by spliting the query column is correct.

df_resized['city'].unique()
```

```python
## Renaming city from Red to Red Deer

df_resized['city'] = df_resized['city'].replace(['Red'], 'Red Deer')
```

```python
## checking operation of renaming city is successful.

df_resized['city'].unique()
```

#### iv) Removing rows where review_id is "__NO_REVIEWS_FOUND__"

```python
df_resized.drop(df_resized[df_resized['review_id'] == '__NO_REVIEWS_FOUND__'].index, inplace=True)
```

```python
df_resized.shape
```

#### v) Inspect names of stores by checking the unique names to identify those that are not african stores.  Prehistoria Museum & SkullStore Oddity Shop, Browns Socialhouse Lethbridge West, Kiim's Market, will be removed because they are not African stores.

```python
df_resized['name'].unique()
```

```python
df_resized.drop(df_resized[df_resized['name'] == 'Prehistoria Museum & SkullStore Oddity Shop'].index, inplace=True)
df_resized.drop(df_resized[df_resized['name'] == 'Browns Socialhouse Lethbridge West'].index, inplace=True)
df_resized.drop(df_resized[df_resized['name'] == "Kiim's Market"].index, inplace=True)
```

```python
df_resized.shape
```

#### vi) Checking for duplicate rows of data.

```python
duplicate_rows_df = df_resized[df_resized.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
## removing dupliacte rows.

df_resized_dedup = df_resized.drop_duplicates().reset_index(drop = True)
```

```python
## checking for number of rows remaining.

df_resized_dedup.shape
```

```python
## checking for number of stores.

df_resized_dedup['name'].nunique()
```

## 3. Data Export to csv file.

```python
## exporting the clean data frame for EDA & visualization

df_resized_dedup.to_csv('outscraper2_data4_EDA.csv')
```

```python
## creating a new dataframe from the clean data frame that contains only 2 columns: 'name', 'review_text'

df_outscraper = df_resized_dedup[['name', 'review_text']]
```
